<b>Question 2</b><br>
Plot a line plot showing the sales trend in company 1 and 2:<br>
days = [1,2,3,4,5,6,7] #days of d week<br>
sales_1 = [160,150,140,145,175,165,180] #sales of company1<br>
sales_2 = [70,90,160,150,140,145,175]  #sales of company2<br>

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

```python
plt.plot(days,sales_1,"bo-")
plt.plot(days,sales_2,"ro-")
plt.title("Days vs Sales")
plt.legend(["sales1","sales2"])
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```

<b>Question 3</b><br>
Create a 3 by 3 subplots:<br>
multiple plots<br>
x = [1,2,3,4]<br>
y1 = [4,3,2,1]<br>
y2 = [10,20,30,40]<br>
y3 = [40,30,20,10]<br>
y4 = [1,2,1,2]<br>
y5 = [40,70,90,70]<br>

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
plt.subplot(3,3,1)
plt.plot(x,y1,"r--")
plt.subplot(3,3,2)
plt.xlabel("x")
plt.ylabel("y1")
plt.plot(x,y2,"g--")
plt.subplot(3,3,3)
plt.plot(x,y3,"b--")
plt.subplot(3,3,4)
plt.plot(x,y4,"m--")
plt.subplot(3,3,5)
plt.plot(x,y5,"k--")
plt.show()
```
