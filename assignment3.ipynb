```python
#IMPORTING NECESSARY PACKAGES
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

<b>Question 1</b><br>
Plot using sns.relplot, line plot, box plot on fmri data set

```python
fmri = sns.load_dataset("fmri")
fmri.head(10)
```

```python
fmri.info()
```

```python
print("CATEGORICAL VARIABLES")
print("SUBJECT : ",fmri['subject'].unique())
print("EVENT : ",fmri['event'].unique())
print("REGION : ",fmri['region'].unique())
print("\nNUMERICAL VARIABLES DESCRIPTION")
print("TIMEPOINT : ",fmri['timepoint'].unique())
print(fmri.describe())
```

```python
print('Timepoint vs Signal')
sns.relplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
print('Timepoint vs Signal for Event')
sns.relplot(x="timepoint", y="signal", data=fmri, hue='event')
plt.show()
```

```python
print('Timepoint vs Signal for Subject')
sns.relplot(x="timepoint", y="signal", data=fmri, hue='subject')
plt.show()
```

```python
print('Timepoint vs Signal for Region')
sns.relplot(x="timepoint", y="signal", data=fmri, hue='region')
plt.show()
```

```python
print('Timepoint vs Signal')
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
sns.lineplot(ax=axes[0],x="timepoint", y="signal", data=fmri, hue='event')
axes[0].set_title("Timepoint vs Signal for Event")
sns.lineplot(ax=axes[1],x="timepoint", y="signal", data=fmri, hue='region')
axes[1].set_title("Timepoint vs Signal for Region")
sns.lineplot(ax=axes[2],x="timepoint", y="signal", data=fmri, hue='subject')
axes[2].set_title("Timepoint vs Signal for Subject")
plt.show()
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
print('Timepoint vs Signal')
sns.boxplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
sns.boxplot(x="timepoint", y="signal", data=fmri, hue='event').set_title('Timepoint vs Signal for Event')
plt.show()
```

```python
sns.boxplot(x="timepoint", y="signal", data=fmri, hue='subject').set_title('Timepoint vs Signal for Subject')
plt.show()
```

```python
sns.boxplot(x="timepoint", y="signal", data=fmri, hue='region').set_title('Timepoint vs Signal for Region')
plt.show()
```
